```python
# Carga de los datos originales
df = pd.read_csv("cal_mad_01.csv", sep=',')
df

```

```python
#Nuevo dataset con la columna de sexo
ndf = df
gender = ''

# Nueva columna a ndf
for index, row in ndf.iterrows():
    name = row["Nombre"]
    gender = d.get_gender(row["Nombre"].split()[-1])
    ndf.loc[index,'sexo']=gender
    print(f"{name} es del género {gender}")

print(ndf['sexo'])
```

```python
print(ndf.sexo.loc[ndf['Nombre']=='Zepeda Cepeda Jesús'])
```

```python
# Revisión del resultado
xdf = ndf[['Nombre','sexo']].loc[ndf['sexo'].isin({'mostly_male'})]
print(xdf[['Nombre','sexo']])
```

```python
#Convertimos mostly_female y mostly_male a female y male
# Replace multiple values
ndf = ndf.replace(['mostly_female','mostly_male'],['male', 'female'])
print(ndf)
```

```python
# Revisión del resultado
xdf = ndf[['Nombre','sexo']].loc[ndf['sexo'].isin({'unknown'})]
print(xdf[['Nombre','sexo']])
```
